```python
#Define an array for samples per each sample size 
samples_10 = np.zeros(10000) 
samples_20 = np.zeros(10000) 
samples_30 = np.zeros(10000) 
samples_40 = np.zeros(10000) 
samples_50 = np.zeros(10000)
#Defining mu and sigma
mu=(4+5+6+7)/4
print(mu)
n=5
p=1/4
sigma=np.sqrt(n*p)
print(sigma)
```

```python
#Defining parameters 
beta = 1
#Creating for loop for sample size 10
for x in range(20000):
    sample = np.random.exponential(beta, 10)
    x_bar_10 = sample.mean()
    samples_10[x] = np.sqrt(10)*(x_bar_10-1)
    #Creating for loop for sample size 20
for a in range(20000): 
    sample = np.random.exponential(beta, 20)
    x_bar_20 = sample.mean()
    samples_20[a] = np.sqrt(20)*(x_bar_20-1)
#Creating for loop for sample size 30
for b in range(20000):
    sample = np.random.exponential(beta, 30)
    x_bar_30 = sample.mean()
    samples_30[b] = np.sqrt(30)*(x_bar_30-1)
#Creating for loop for sample size 50
for c in range(20000):
    sample = np.random.exponential(beta, 50)
    x_bar_50 = sample.mean()
    samples_50[c] = np.sqrt(50)*(x_bar_50-1)
```

```python
#Combining arrays into a dataframe to plot
df = (samples_10,samples_20,samples_30,samples_50)

#Plotting Distributions
plt.figure(figsize=(20,10))
#sns.distplot(samples_10)
#sns.distplot(samples_20)
#sns.distplot(samples_30)
sns.distplot(samples_50) 
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
# Text: the following histogram represents sample size 50.
```
